```python
# Gaussian lowpass filter
def GLPF(I, D0):
    U,V=dtfuv(I.shape[0],I.shape[1])
    D = IP.np.sqrt(U**2+V**2)
    filter=IP.np.exp(-D**2/(2*(D0**2)))
    If = fftshift(fft2(I))
    filterf = fftshift(fft2(I))
    Ic = If * filterf
    return fftshift(ifft2(Ic))
```

```python
# Uniform noise filter
def uniform_noise(I, alpha=10):
    noise = randn(I.shape[0], I.shape[1])
    return noise*(I.mean()*(alpha/100))+I
```

```python
# Generate "broken" image
Ia = GLPF(I, 40)
Ib = uniform_noise(Ia, 20)
IP.subplot(1,2,1);imshow(I, aspect="equal", cmap="gray")
IP.subplot(1,2,2);imshow(Ib, aspect="equal", cmap="gray")
```

```python
# Deblurring given known gaussian blur size 40
```

```python
# "inverse" Gaussian lowpass filter
def iGLPF(I, D0):
    U,V=dtfuv(I.shape[0],I.shape[1])
    D = IP.np.sqrt(U**2+V**2)
    filter=IP.np.exp(-D**2/(2*(D0**2)))
    If = fft2(I,[I.shape[0],I.shape[1]])
    filterf = np.abs(fft2(filter,[I.shape[0],I.shape[1]]))
    epsilon = 1e-10
    deblurred = If / (filterf+epsilon)
    return np.abs(ifft2(deblurred))

#     If = fftfilt(I, filter)
#     If = If-If.min()
#     If =(If/If.max())*256
#     return If
```

```python
deblur_a = iGLPF(Ia, 40) # only blur
deblur_b = iGLPF(Ib, 40) # blur and random uniform noise
IP.subplot(1,2,1);imshow(deblur_a, aspect="equal", cmap="gray")
IP.subplot(1,2,2);imshow(deblur_b, aspect="equal", cmap="gray")
```
